### 실습 과제

1. 800x400(가로x세로) 크기의 `RGB=[0, 102, 0]` space를 완성하세요.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

space = np.zeros([400, 800, 3], dtype=np.uint8)
space[:, :, :] = [0, 102, 0]
plt.imshow(space)
```

2. `cv2.line()` 함수를 이용하여 [100, 100]에서 [600, 300]까지 흰색 선을 그려보세요.

```python
space = np.zeros([400, 800, 3], dtype=np.uint8)
space[:, :, :] = [0, 102, 0]
cv2.line(space, [100, 100], [600, 300], [255, 255, 255], 3)
plt.imshow(space)
plt.show()
```

3. `cv2.circle()` 함수를 이용해서 중심 [400, 200]에서 반지름이 100인 흰색 원을 그려보세요.

```python
space = np.zeros([400, 800, 3], dtype=np.uint8)
space[:, :, :] = [0, 102, 0]
space = cv2.circle(space, [400, 200], 100, [255, 255, 255], 3)
plt.imshow(space)
plt.show()
```

4. `cv.rectangle()` 함수를 이용하여 좌상단 [100, 100], 우하단 [700, 300]의 흰색 직사각형을 그려보세요.

```python
space = np.zeros([400, 800, 3], dtype=np.uint8)
space[:, :, :] = [0, 102, 0]
space = cv2.rectangle(space, [100, 100], [700, 300], [255, 255, 255], 3)
plt.imshow(space)
plt.show()
```
