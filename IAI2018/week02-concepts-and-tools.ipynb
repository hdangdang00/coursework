```python
r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]
```

```python
plt.hist(r.reshape(-1), range=(0,255), bins=255)
```

```python
# r, g, b의 히스토그램을 임의로 넓게 펼침
rr = cv2.equalizeHist(r)
gg = cv2.equalizeHist(g)
bb = cv2.equalizeHist(b)
```

```python
# r과 rr의 히스토그램을 비교
plt.hist(r.reshape(-1), range=(0,255), bins=255)
plt.hist(rr.reshape(-1), range=(0,255), bins=255)
plt.show()
```

```python
plt.imshow(img) # 원본 이미지
```

```python
img2 = np.stack([rr, gg, bb], axis=-1)
plt.imshow(img2)
```

### 실습 과제

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./img/overexposed.jpg') # OpenCV는 BGR 순서로 이미지를 읽어옴
plt.imshow(img)
```

```python
print(img.shape)

r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]

plt.hist(r.reshape(-1), range=(0, 255), bins=255)
```

```python
rr = cv2.equalizeHist(r)
gg = cv2.equalizeHist(g)
bb = cv2.equalizeHist(b)

plt.hist(r.reshape(-1), bins=255, range=[0, 255], label='befor')
plt.hist(rr.reshape(-1), bins=255, range=[0, 255], label='after')
plt.legend()
```

```python
img2 = np.stack([rr, gg, bb], axis=-1)
plt.imshow(np.concatenate([img, img2], axis=1))
```
