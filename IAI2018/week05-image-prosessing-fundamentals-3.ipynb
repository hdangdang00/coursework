#### 카메라 출력 → 실시간 사이즈 변환 → 합쳐서 출력
* `np.hstack()` 함수를 이용하여 출력 이미지 합치기
* `np.hstack()`과 `np.vstack()`은 합치는 배열의 사이즈가 동일 해야 함

* opencv에서는 항상 **width, height** 순서!!

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
while cv2.waitKey(33) < 0:
    _, frame = capture.read()
    frame1 = cv2.resize(frame, [300, 150])
    frame2 = cv2.resize(frame, [500, 150])
    frame3 = np.hstack([frame1, frame2])
    cv2.imshow('video1+2', frame3)

capture.release()
cv2.destroyAllWindows()
```

#### fx, fy 사용하기

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
while cv2.waitKey(33) < 0:
    _, frame = capture.read()
    frame1 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame2 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame3 = cv2.resize(frame, dsize=None, fx=2, fy=2)
    frame4 = np.vstack([frame1, frame2])
    frame5 = np.hstack([frame4, frame3])
    cv2.imshow('video1+2+3', frame5)

capture.release()
cv2.destroyAllWindows()
```

#### 플립한 동영상 추가해 보기

#### 회전한 동영상 추가해 보기

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
while cv2.waitKey(33) < 0:
    _, frame = capture.read()
    frame1 = cv2.resize(frame, dsize=None, fx=0.5, fy=0.5)
    frame2 = cv2.resize(frame, dsize=None, fx=0.5, fy=0.5)
    frame3 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    h, w, _ = frame.shape # 배열에서 가져왔기 때문에 (height, width)
    matrix = cv2.getRotationMatrix2D([w/2, h/2], 90, 1) # opencv는 (width, height)
    frame3 = cv2.warpAffine(frame3, matrix, [w, h])
    frame4 = np.vstack([frame1, frame2])
    frame5 = np.hstack([frame4, frame3])
    cv2.imshow('v1+2+3', frame5)

capture.release()
cv2.destroyAllWindows()
```

#### 회전한 동영상에 borderValue 추가해 보기

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
while cv2.waitKey(33) < 0:
    _, frame = capture.read()
    frame1 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame2 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame3 = cv2.resize(frame, dsize=None, fx=2, fy=2)
    h, w, _ = frame3.shape
    matrix = cv2.getRotationMatrix2D([w/2, h/2], 30, 1)
    frame3 = cv2.warpAffine(frame3, matrix, [w, h], borderValue=[123, 123, 0])
    frame4 = np.vstack([frame1, frame2])
    frame5 = np.hstack([frame4, frame3])
    cv2.imshow('v1+2+3', frame5)

capture.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
while cv2.waitKey(33) < 0:
    _, frame = capture.read()
    frame1 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame2 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame3 = cv2.resize(frame, dsize=None, fx=2, fy=2)
    h, w, _ = frame3.shape
    matrix = cv2.getRotationMatrix2D([w/2, h/2], 30, 1)
    frame3 = cv2.warpAffine(frame3, matrix, [w, h], borderValue=[123, 123, 0])
    zf0 = np.zeros([h, w], dtype=np.uint8)
    frame3 = cv2.merge([zf0, zf0, frame3[:, :, 2]])
    frame4 = np.vstack([frame1, frame2])
    frame5 = np.hstack([frame4, frame3])
    cv2.imshow('v1+2+3', frame5)

capture.release()
cv2.destroyAllWindows()
```

#### 이진 변환한 동영상 추가해 보기

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
while cv2.waitKey(33) < 0:
    _, frame = capture.read()
    frame1 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame2 = cv2.resize(frame, dsize=None, fx=1, fy=1)
    frame3 = cv2.resize(frame, dsize=None, fx=2, fy=2)
    h, w, _ = frame3.shape
    matrix = cv2.getRotationMatrix2D([w/2, h/2], 30, 1)
    frame3 = cv2.warpAffine(frame3, matrix, [w, h], borderValue=[123, 123, 0])
    frame3 = cv2.cvtColor(frame3, cv2.COLOR_BGR2GRAY)
    _, frame3 = cv2.threshold(frame3, 100, 255, cv2.THRESH_BINARY)
    frame3 = cv2.cvtColor(frame3, cv2.COLOR_GRAY2BGR) # 변환 안해주면 에러 발생
    frame4 = np.vstack([frame1, frame2])
    frame5 = np.hstack([frame4, frame3])
    cv2.imshow('v1+2+3', frame5)

capture.release()
cv2.destroyAllWindows()
```
