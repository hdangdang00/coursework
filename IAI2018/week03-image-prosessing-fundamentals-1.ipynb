#### 카메라 출력

```python
import cv2

capture = cv2.VideoCapture(0)
capture.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
capture.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

while cv2.waitKey(33) < 0:
    ret, frame = capture.read()
    cv2.imshow('VideoFrame', frame)

capture.release()
cv2.destroyAllWindows()
```

#### wget으로 이미지 다운로드 받아 cv2로 읽기

```python
import wget
import cv2

wget.download('https://raw.githubusercontent.com/tysep16/CV26_1/refs/heads/main/crow.jpeg', './img/')

image = cv2.imread('./img/crow.jpeg')
cv2.imshow('BGR', image)
cv2.waitKey()
cv2.destroyAllWindows()

print(image[:1, :1, :3]) # 첫 번째 픽셀의 BGR 값 출력
```

BGR 채널 분리 - cv2로 읽기

```python
import cv2
image = cv2.imread('./img/crow.jpeg')

b = image[:, :, 0]
g = image[:, :, 1]
r = image[:, :, 2]

cv2.imshow('R', r)
cv2.imshow('G', g)
cv2.imshow('B', b)

cv2.waitKey()
cv2.destroyAllWindows()
```

#### 이미지 출력 - plt로 읽기

```python
import matplotlib.pyplot as plt
image1 = plt.imread('./img/crow.jpeg')
plt.imshow(image1)
```

```python
print(image1[0, 0, :])
```

**정리: OpenCV는 BGR, matplotlib은 RGB로 색상 표현**

RGB 채널 분리 - plt로 읽기

```python
r = image1[:, :, 0]
g = image1[:, :, 1]
b = image1[:, :, 2]

fig, axes = plt.subplots(1, 3, figsize=(8, 4))
axes[0].imshow(r, cmap='gray')
axes[0].axis('off')
axes[1].imshow(g, cmap='gray')
axes[1].axis('off')
axes[2].imshow(b, cmap='gray')
axes[2].axis('off')
plt.show()
```

RGB 색공간으로 표현된 이미지는 `plt.imshow()`로 출력

```python
plt.imshow(image1)
```

cv2로 이미지를 읽었으면 cv2로 출력 (plt로 출력하면 왜곡 발생 (에러 메시지 x))

plt로 이미지를 읽었으면 plt로 출력 (cv2로 출력하면 왜곡 발생 (에러 메시지 x))

```python
import cv2

cv2.imshow('read:plt, show:cv2', image1)
cv2.waitKey()
cv2.destroyAllWindows()
```
