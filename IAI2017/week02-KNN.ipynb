### 오프라인 실습


```python
# 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
len(bream_length)
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
len(smelt_length)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
fish_data
```

```python
# 정답 데이터
fish_target = [1] * 35 + [0] * 14
fish_target
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier() # n_neighbors의 default 값은 5
kn.fit(fish_data, fish_target) # KNN의 기본 학습은 fit(input data, target data) 메서드를 사용
```

```python
kn.score(fish_data, fish_target) # 학습한 데이터로 모델 평가 (0.0 ~ 1.0 사이의 값)
```

```python
kn.predict([[30, 600]]) # predict 메서드는 중첩 리스트로 입력해야 함
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
```

```python
kn49.fit(fish_data, fish_target)
```

```python
kn49.score(fish_data, fish_target) # 전체 데이터 중 도미 데이터의 비율이 약 0.7이므로 n_neighbors=49로 설정하면 모든 데이터를 도미로 분류하여 0.7의 정확도가 나옴
```

```python
kn49.predict([[1000000000, 1000000000]]) # 어떤 입력 값을 넣던 예측값이 도미로 나오게 됨
```

### 실습 과제


1. 붓꽃 꽃받침 리스트 데이터를 이용하여 scatter 플롯을 출력하세요.


```python
setona_sepal_length = [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 4.6, 5.0, 4.4, 4.9]
setona_sepal_width = [3.5, 3.0, 3.2, 3.1, 3.6, 3.9, 3.4, 3.4, 2.9, 3.1]
versicolor_sepal_length = [5.5, 6.1, 5.8, 5.0, 5.6, 5.7, 5.7, 6.2, 5.1, 5.7]
versicolor_sepal_width = [2.6, 3.0, 2.6, 2.3, 2.7, 3.0, 2.9, 2.9, 2.5, 2.8]

plt.scatter(setona_sepal_length, setona_sepal_width)
plt.scatter(versicolor_sepal_length, versicolor_sepal_width)
plt.xlabel("length")
plt.ylabel("width")
plt.show()
```

2. setona와 versicolor의 length와 width를 하나의 리스트로 각각 합친 후, zip() 함수를 사용하여 [[length, width], ...] 형태의 중첩 리스트로 만들어 보세요.


```python
iris_length = setona_sepal_length + versicolor_sepal_length
iris_width = setona_sepal_width + versicolor_sepal_width
iris_data = [[l, w] for l, w in zip(iris_length, iris_width)]
print(iris_data)
```

3. target 데이터를 만든 후 sklearn.neighbors의 KNeighborsClassifier 클래스를 이용하여 모델을 생성하고 평가해 보세요.


```python
iris_target = [0] * 10 + [1] * 10

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(iris_data, iris_target)
kn.score(iris_data, iris_target)
```

4. 꽃받침의 길이(length)와 너비(width)가 각각 5, 3인 꽃은 setona일까? versicolor일까?
   단, prediction 값이 0이면 setona, 1이면 versicolor


```python
kn.predict([[5.0, 3.0]])
```
