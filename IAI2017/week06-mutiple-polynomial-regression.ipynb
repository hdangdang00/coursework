### 실습 과제

1. fifa25.csv 데이터로 데이터프레임을 생성 후, 데이터 프레임의 'PAC', 'SHO', 'PAS', 'DRI', 'DEF' 열을 input_data로, 'OVR' 열을 output_data로 지정하세요.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

df = pd.read_csv('https://raw.githubusercontent.com/tysep16/DV25_1/refs/heads/main/fifa25.csv')
input_data = df[['PAC', 'SHO', 'PAS', 'DRI', 'DEF']]
output_data = df['OVR']
```

2. input_data와 output_data를 train dataset, test dataset으로 구분 후, polynomial feature로 확장한 후 정규화를 하세요.

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, random_state=42)
poly = PolynomialFeatures(include_bias=False)
poly.fit(x_train)
train_poly = poly.transform(x_train)
test_poly = poly.transform(x_test)
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

3. Linear Regression 모델을 생성 후 score 평가해보세요.

```python
lr = LinearRegression()
lr.fit(train_scaled, y_train)
lr.score(test_scaled, y_test)
```
