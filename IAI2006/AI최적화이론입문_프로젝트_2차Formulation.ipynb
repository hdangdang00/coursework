```python
from gurobipy import Model, GRB

foods_list = [
    {
        "name": "삶은계란",
        "weight": 50,
        "cal": 71,
        "protein": 6,
        "fat": 5,
        "carb": 0.5,
        "price": 500,
        "category": "SIDE_DISH",
    },
    {
        "name": "잡채",
        "weight": 75,
        "cal": 257,
        "protein": 2,
        "fat": 5,
        "carb": 51,
        "price": 1495,
        "category": "MAIN_DISH",
    },
    {
        "name": "사과",
        "weight": 150,
        "cal": 63.8,
        "protein": 0.5,
        "fat": 0.2,
        "carb": 15,
        "price": 3400,
        "category": "OTHER",
    },
    {
        "name": "제육볶음",
        "weight": 200,
        "cal": 430,
        "protein": 25,
        "fat": 30,
        "carb": 15,
        "price": 3500,
        "category": "MAIN_DISH",
    },
    {
        "name": "두부",
        "weight": 100,
        "cal": 92.9,
        "protein": 9.4,
        "fat": 5.3,
        "carb": 1.9,
        "price": 500,
        "category": "SIDE_DISH",
    },
    {
        "name": "물만두",
        "weight": 370,
        "cal": 696,
        "protein": 29,
        "fat": 32,
        "carb": 73,
        "price": 4500,
        "category": "SIDE_DISH",
    },
    {
        "name": "고등어 구이",
        "weight": 100,
        "cal": 272,
        "protein": 24,
        "fat": 19,
        "carb": 1,
        "price": 2500,
        "category": "MAIN_DISH",
    },
    {
        "name": "돼지고기 김치찌개",
        "weight": 460,
        "cal": 240,
        "protein": 24,
        "fat": 8,
        "carb": 18,
        "price": 7410,
        "category": "SOUP",
    },
    {
        "name": "고추참치",
        "weight": 100,
        "cal": 142,
        "protein": 10,
        "fat": 6,
        "carb": 12,
        "price": 3300,
        "category": "SIDE_DISH",
    },
    {
        "name": "차돌된장찌개",
        "weight": 460,
        "cal": 280,
        "protein": 17,
        "fat": 12,
        "carb": 26,
        "price": 7000,
        "category": "SOUP",
    },
    {
        "name": "카레",
        "weight": 200,
        "cal": 171,
        "protein": 5,
        "fat": 7,
        "carb": 22,
        "price": 1980,
        "category": "SOUP",
    },
    {
        "name": "쌀밥",
        "weight": 210,
        "cal": 297,
        "protein": 5,
        "fat": 1,
        "carb": 67,
        "price": 1000,
        "category": "RICE",
    },
    {
        "name": "참치마요 삼각김밥",
        "weight": 110,
        "cal": 180,
        "protein": 5,
        "fat": 8,
        "carb": 22,
        "price": 1700,
        "category": "RICE",
    },
    {
        "name": "삶은감자",
        "weight": 200,
        "cal": 176.76,
        "protein": 3.72,
        "fat": 0.2,
        "carb": 40.02,
        "price": 750,
        "category": "OTHER",
    },
    {
        "name": "낫또",
        "weight": 40,
        "cal": 89,
        "protein": 7,
        "fat": 5,
        "carb": 4,
        "price": 700,
        "category": "SIDE_DISH",
    },
    {
        "name": "계란말이",
        "weight": 100,
        "cal": 133,
        "protein": 10,
        "fat": 9,
        "carb": 3,
        "price": 1000,
        "category": "SIDE_DISH",
    },
    {
        "name": "떡갈비",
        "weight": 100,
        "cal": 318,
        "protein": 12,
        "fat": 26,
        "carb": 9,
        "price": 2200,
        "category": "MAIN_DISH",
    },
]

# 상수 정의
MIN_CAL = 2000  # 하루 최소 섭취 칼로리
MAX_CAL = 2500  # 하루 최대 섭취 칼로리
CARB_CAL = 4  # 탄수화물 1g당 칼로리
PROTEIN_CAL = 4  # 단백질 1g당 칼로리
FAT_CAL = 9  # 지방 1g당 칼로리

# 데이터 전처리
food_names = [f["name"] for f in foods_list]
foods_dict = {f["name"]: f for f in foods_list}
categories = list(set(f["category"] for f in foods_list))

# 모델 생성
m = Model("diet_2nd_formulation")

# 결정 변수 정의
b = m.addVars(food_names, name="breakfast", vtype=GRB.INTEGER)
l = m.addVars(food_names, name="lunch", vtype=GRB.INTEGER)
d = m.addVars(food_names, name="dinner", vtype=GRB.INTEGER)

# 결정변수: 각 끼니별 카테고리 선택 여부
b_cat = m.addVars(categories, name="breakfast_category_selected", vtype=GRB.BINARY)
l_cat = m.addVars(categories, name="lunch_category_selected", vtype=GRB.BINARY)
d_cat = m.addVars(categories, name="dinner_category_selected", vtype=GRB.BINARY)

# 목적 함수 정의 - 총 단백질 섭취량 최대화
m.setObjective(
    sum(foods_dict[f]["protein"] * (b[f] + l[f] + d[f]) for f in food_names),
    GRB.MAXIMIZE,
)

totals = {
    "cal": sum(foods_dict[f]["cal"] * (b[f] + l[f] + d[f]) for f in food_names),
    "carb": sum(foods_dict[f]["carb"] * (b[f] + l[f] + d[f]) for f in food_names),
    "protein": sum(foods_dict[f]["protein"] * (b[f] + l[f] + d[f]) for f in food_names),
    "fat": sum(foods_dict[f]["fat"] * (b[f] + l[f] + d[f]) for f in food_names),
    "price": sum(foods_dict[f]["price"] * (b[f] + l[f] + d[f]) for f in food_names),
}

# 공통 제약식 (Common Constraints)
# c-1. 하루 최소 칼로리: 2000 kcal 이상
m.addConstr(totals["cal"] >= MIN_CAL, "c_min_calories")

# c-2. 탄수화물 비율: 총 칼로리의 55% 이상
m.addConstr(totals["carb"] * CARB_CAL >= totals["cal"] * 0.55, "c_carb_min_ratio")

# c-3. 탄수화물 비율: 총 칼로리의 65% 이하
m.addConstr(totals["carb"] * CARB_CAL <= totals["cal"] * 0.65, "c_carb_max_ratio")

# c-4. 단백질 비율: 총 칼로리의 7% 이상
m.addConstr(
    totals["protein"] * PROTEIN_CAL >= totals["cal"] * 0.07, "c_protein_min_ratio"
)

# c-5.단백질 비율: 총 칼로리의 20% 이하
m.addConstr(
    totals["protein"] * PROTEIN_CAL <= totals["cal"] * 0.20, "c_protein_max_ratio"
)

# c-6. 지방 비율: 총 칼로리의 15% 이상
m.addConstr(totals["fat"] * FAT_CAL >= totals["cal"] * 0.15, "c_fat_min_ratio")

# c-7. 지방 비율: 총 칼로리의 30% 이하
m.addConstr(totals["fat"] * FAT_CAL <= totals["cal"] * 0.30, "c_fat_max_ratio")

# 개인별 제약식 (Personal Constraints)
# p-1. 하루 최대 칼로리: 2500 kcal 이하
m.addConstr(totals["cal"] <= MAX_CAL, "p_max_cal")

# p-2. 각 끼니에 최소 1인분 이상 섭취
m.addConstr(sum(b[f] for f in food_names) >= 1, "p_breakfast_at_least_one")
m.addConstr(sum(l[f] for f in food_names) >= 1, "p_lunch_at_least_one")
m.addConstr(sum(d[f] for f in food_names) >= 1, "p_dinner_at_least_one")

# 적응적 제약식 (Adaptive Constraints)
# a-1. 각 음식은 끼니 당 최대 1인분까지 섭취 가능
for f in food_names:
    m.addConstr(b[f] <= 1, f"a_breakfast_{f}_max_1")
    m.addConstr(l[f] <= 1, f"a_lunch_{f}_max_1")
    m.addConstr(d[f] <= 1, f"a_dinner_{f}_max_1")

# a-2. 카테고리 선택 연결 제약
# 음식을 섭취하면 해당 카테고리가 자동으로 선택됨 (Big-M 기법)
for c in categories:
    foods_in_category = [f for f in food_names if foods_dict[f]["category"] == c]

    if len(foods_in_category) > 0:
        N_c = len(foods_in_category)  # Big M (해당 카테고리 최대 섭취 가능 인분 합)
        m.addConstr(
            sum(b[f] for f in foods_in_category) <= N_c * b_cat[c],
            f"a_breakfast_cat_active_max_{c}",
        )
        m.addConstr(
            sum(b[f] for f in foods_in_category) >= b_cat[c],
            f"a_breakfast_cat_active_min_{c}",
        )

        m.addConstr(
            sum(l[f] for f in foods_in_category) <= N_c * l_cat[c],
            f"a_lunch_cat_active_max_{c}",
        )
        m.addConstr(
            sum(l[f] for f in foods_in_category) >= l_cat[c],
            f"a_lunch_cat_active_min_{c}",
        )

        m.addConstr(
            sum(d[f] for f in foods_in_category) <= N_c * d_cat[c],
            f"a_dinner_cat_active_max_{c}",
        )
        m.addConstr(
            sum(d[f] for f in foods_in_category) >= d_cat[c],
            f"a_dinner_cat_active_min_{c}",
        )

# a-3. 끼니별 카테고리 다양성 제한
# 각 끼니에서 최대 3개 카테고리까지만 선택
m.addConstr(sum(b_cat[c] for c in categories) <= 3, "a_breakfast_max_3_categories")
m.addConstr(sum(l_cat[c] for c in categories) <= 3, "a_lunch_max_3_categories")
m.addConstr(sum(d_cat[c] for c in categories) <= 3, "a_dinner_max_3_categories")

# a-4. RICE 카테고리 필수 섭취
rice_foods = [f for f in food_names if foods_dict[f]["category"] == "RICE"]

m.addConstr(sum(b[f] for f in rice_foods) == 1, "a_breakfast_rice_required")
m.addConstr(sum(l[f] for f in rice_foods) == 1, "a_lunch_rice_required")
m.addConstr(sum(d[f] for f in rice_foods) == 1, "a_dinner_rice_required")

# a-5. SOUP 카테고리 섭취량 제한
soup_foods = [f for f in food_names if foods_dict[f]["category"] == "SOUP"]

m.addConstr(sum(b[f] for f in soup_foods) <= 1, "a_breakfast_max_1_soup")
m.addConstr(sum(l[f] for f in soup_foods) <= 1, "a_lunch_max_1_soup")
m.addConstr(sum(d[f] for f in soup_foods) <= 1, "a_dinner_max_1_soup")

# a-6. 끼니별 칼로리 배분
# 각 끼니당 섭취 칼로리는 섭취 최대 칼로리의 아침 30%, 점심 40%, 저녁 30%의 비율로 배분
b_cal = sum(foods_dict[f]["cal"] * b[f] for f in food_names)
l_cal = sum(foods_dict[f]["cal"] * l[f] for f in food_names)
d_cal = sum(foods_dict[f]["cal"] * d[f] for f in food_names)

m.addConstr(b_cal <= MAX_CAL * 0.3, "a_breakfast_cal_ratio")
m.addConstr(l_cal <= MAX_CAL * 0.4, "a_lunch_cal_ratio")
m.addConstr(d_cal <= MAX_CAL * 0.3, "a_dinner_cal_ratio")

# a-7. 카테고리별 하루 섭취 제한 (RICE 제외)
# 식단 다양성 확보를 위해 RICE를 제외한 카테고리는 하루 최대 1회만 섭취
non_rice_categories = [c for c in categories if c != "RICE"]

for c in non_rice_categories:
    m.addConstr(b_cat[c] + l_cat[c] + d_cat[c] <= 1, f"adaptive_max_one_per_day_{c}")

# 최적화 실행
m.optimize()

if m.status == GRB.OPTIMAL:
    print("\n" + "=" * 70)
    print("  My 1 Day Diet Program: 2nd Formulation")
    print("=" * 70)
    print(f"\n  목적함수 값 (총 단백질 섭취량): {m.objVal:.2f}g\n")
    print("-" * 70 + "\n")

    # 아침 식사
    selected_b_cats = [c for c in categories if b_cat[c].X > 0.5]
    print("=" * 70)
    print(f"  아침 식사 ({len(selected_b_cats)}개 카테고리)")
    print("=" * 70)
    print(f"  선택된 카테고리: {', '.join(selected_b_cats)}")
    print("-" * 70)
    b_cal_total = 0
    b_protein_total = 0
    b_price_total = 0
    for f in food_names:
        if b[f].X > 0.0001:
            print(f"  • {f}: {b[f].X:.0f}인분")
            b_cal_total += foods_dict[f]["cal"] * b[f].X
            b_protein_total += foods_dict[f]["protein"] * b[f].X
            b_price_total += foods_dict[f]["price"] * b[f].X
    print("-" * 70)
    print(
        f"  칼로리: {b_cal_total:.2f}kcal | 단백질: {b_protein_total:.2f}g | 비용: {b_price_total:,.0f}원\n"
    )

    # 점심 식사
    selected_l_cats = [c for c in categories if l_cat[c].X > 0.5]
    print("=" * 70)
    print(f"  점심 식사 ({len(selected_l_cats)}개 카테고리)")
    print("=" * 70)
    print(f"  선택된 카테고리: {', '.join(selected_l_cats)}")
    print("-" * 70)
    l_cal_total = 0
    l_protein_total = 0
    l_price_total = 0
    for f in food_names:
        if l[f].X > 0.0001:
            print(f"  • {f}: {l[f].X:.0f}인분")
            l_cal_total += foods_dict[f]["cal"] * l[f].X
            l_protein_total += foods_dict[f]["protein"] * l[f].X
            l_price_total += foods_dict[f]["price"] * l[f].X
    print("-" * 70)
    print(
        f"  칼로리: {l_cal_total:.2f}kcal | 단백질: {l_protein_total:.2f}g | 비용: {l_price_total:,.0f}원\n"
    )

    # 저녁 식사
    selected_d_cats = [c for c in categories if d_cat[c].X > 0.5]
    print("=" * 70)
    print(f"  저녁 식사 ({len(selected_d_cats)}개 카테고리)")
    print("=" * 70)
    print(f"  선택된 카테고리: {', '.join(selected_d_cats)}")
    print("-" * 70)
    d_cal_total = 0
    d_protein_total = 0
    d_price_total = 0
    for f in food_names:
        if d[f].X > 0.0001:
            print(f"  • {f}: {d[f].X:.0f}인분")
            d_cal_total += foods_dict[f]["cal"] * d[f].X
            d_protein_total += foods_dict[f]["protein"] * d[f].X
            d_price_total += foods_dict[f]["price"] * d[f].X
    print("-" * 70)
    print(
        f"  칼로리: {d_cal_total:.2f}kcal | 단백질: {d_protein_total:.2f}g | 비용: {d_price_total:,.0f}원\n"
    )

    # 전체 합계
    total_cal = b_cal_total + l_cal_total + d_cal_total
    total_protein = b_protein_total + l_protein_total + d_protein_total
    total_price = b_price_total + l_price_total + d_price_total

    print("=" * 70)
    print("  식단 최적화 결과")
    print("=" * 70 + "\n")
    print(f"  • 총 칼로리: {total_cal:.2f}kcal")
    print(f"  • 총 단백질: {total_protein:.2f}g")
    print(f"  • 총 비용: {total_price:,.0f}원")
    print("\n" + "-" * 70)
    # 영양소 비율 검증
    carb_cal = totals["carb"].getValue() * CARB_CAL
    protein_cal = totals["protein"].getValue() * PROTEIN_CAL
    fat_cal = totals["fat"].getValue() * FAT_CAL

else:
    print(f"\n최적해를 찾을 수 없습니다. (Status: {m.status})")
```
